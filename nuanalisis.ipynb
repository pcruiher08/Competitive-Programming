```python
import matplotlib.pyplot as plt
import numpy as np

def compound_interest(principal, rate, time, compounding_frequency):
    n = compounding_frequency
    r = rate / 100 / n
    t = time * n

    return principal * (1 + r)**t

def investment_growth_over_time(principal, rate, time, compounding_frequency, daily_contributions=0):
    n = compounding_frequency
    r = rate / 100 / n
    t = time * n

    time_points = np.arange(0, t + 1, 1)
    values = [compound_interest(principal + i * daily_contributions, rate, i/n, n) for i in time_points]

    return time_points, values

def investment_growth_over_time_custom(principal, rate, time, compounding_frequency, daily_contributions = 0):
    n = compounding_frequency
    r = rate / 100 / n
    t = time * n

    time_points = np.arange(0, t + 1, 1)
    calc = principal
    #values = [compound_interest(principal, rate, i/n, n) for i in time_points]
    delta = [0]
    values = [0]
    for i in range(compounding_frequency * time + 1):
        calc += calc * rate / 100 / 365
        values.append(calc - principal)
        if(i > 0):
            delta.append(values[i] - values[i - 1])
    

    return time_points, delta

# Set initial parameters
rate = 15  # Constant annual interest rate
time = 1 # Number of years
daily_contributions = 1  # Daily contributions to the account

# Define initial investment amounts
initial_investments = [1000, 1500, 2000, 2500, 5000]
#initial_investments = [1000]

# Plot the graph for each initial investment amount
plt.figure(figsize=(10, 6))

results = []
for principal in initial_investments:
    time_points, values = investment_growth_over_time_custom(principal, rate, time, 30, daily_contributions)
    results.append([list(time_points), values])
    plt.plot(time_points, values, label=f'${principal}')

print(results)

plt.title('Investment Growth Over Time with Different Initial Investment Amounts (Constant Rate)')
plt.xlabel('Days')

plt.ylabel('Investment Value ($)')
plt.legend()
plt.legend(title='Initial Investment')

plt.grid(True)
plt.show()
```
